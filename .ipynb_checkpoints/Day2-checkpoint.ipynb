```python
data = {'price': [100, 200, 150, 300], 'quantity':[5,2,1,3]}
df1 = pd.DataFrame(data)
# Apply a lambda function to calculate discounted price
df1['new_price'] = df1['price'].apply(lambda x: x * 0.9)
df1
```

```python
def classify_number(number):
    if number < 3:
        return 'small'
    elif number >= 3 and number < 5:
        return 'medium'
    else:
        return 'large'
```

```python
classify_number(300)
```

```python
df1['quantity_classify'] = df1['quantity'].apply(lambda x: classify_number(x))
df1
```

```python
df1['new_price_classify'] = df1['price'].apply(classify_number)
df1
```

```python
def new_classify_number(number, lower, upper):
    if number < lower:
        return 'small'
    elif number >= lower and number < upper:
        return 'medium'
    else:
        return 'large'
```

```python
new_classify_number(30,25,60)
```

```python
df1['quantity_classify'] = df1['quantity'].apply(lambda x: new_classify_number(x,2,5))
df1
```

```python
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df2 = pd.DataFrame(data)

condition1 = df2['A'] < 3
function1 = lambda x: x * 2
condition2 = df2['A'] >= 3
function2 = lambda x: x + 5

df2['C'] = np.where(condition1, function1(df2['B']), function2(df2['B']))

print(df2)
```

```python
data = {'A': [1, 2, 3, 4, 5], 'B': [10, 20, 30, 40, 50]}
df3 = pd.DataFrame(data)

# Define the conditions and corresponding functions/logic
condition1 = df3['A'] < 3
def function(x,y):
    return x/y
    
condition2 = df3['A'] >= 3
function2 = lambda x: x + 5

# Apply the functions/logic based on conditions and create the new column
df3['C'] = np.where(condition1,function(df3['B'],df3['A']),function2(df3['B']))

print(df3)
```
