```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Revenue_by_actors.csv')
df = df.head(10)
```

```python
plt.figure(figsize=(16,10))
plt.bar(df.actor, df.total_revenue)
plt.title("Top 10 ranked actors with most revenue")
plt.xlabel("Actors")
plt.ylabel("Revenue")
plt.xticks(rotation=30, ha='right')
plt.show()
```
