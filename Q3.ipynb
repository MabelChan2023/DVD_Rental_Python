```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Revenue_by_country.csv')
df.sort_values('total_revenue', inplace=True)
```

```python
country = df['country']
revenue = df['total_revenue']

plt.figure(figsize=(16,28))
plt.barh(country, revenue)
plt.title("Total revenue by countries")
plt.xlabel("Revenue")
plt.ylabel("Countries")
plt.show()
```

```python
country = df['country']
revenue = df['total_revenue']

plt.figure(figsize=(20,20))
plt.pie(revenue, labels = country, autopct='%1.2f%%', pctdistance = 0.9, rotatelabels = True)
plt.title("Total revenue by countries")
plt.legend(country)
plt.show()
```
