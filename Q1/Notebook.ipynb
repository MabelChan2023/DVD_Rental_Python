```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Daily_revenue_by_countries.csv')
```

```python
plt.figure(figsize=(10,6))
plt.title("Time-series plot of daily total revenue for Canada, France, and United States")
plt.xlabel("Date")
plt.ylabel("Revenue")
df.groupby('country')['total_revenue'].plot(legend='True')
plt.show()
```
